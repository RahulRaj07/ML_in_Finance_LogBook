<html> <h3 style="color:blue;"> Correlation Matrix and Heat Map </h3> </html>

..........................................................

```python
data.info()
```

..........................................................

```python
data2 = data.drop(['State'],axis=1)
```

..........................................................

```python
d = {"No": 0,"Yes":1}
data2 = data2.replace({'International plan': d})
data2 = data2.replace({'Voice mail plan': d})
```

..........................................................

```python
data2['Churn'] = data2['Churn'].astype('int8')
```

..........................................................

```python
data2.info()
```

```python
# Now we have only digital data, therefore we can build a correlation matrix
```

..........................................................

```python
correlation_matrix = data2.corr()
correlation_matrix
```

..........................................................

```python
plt.figure(figsize=(12,6))
sns.heatmap(correlation_matrix, annot= True, cmap= 'coolwarm', linewidths=0.5, fmt='.2f')
plt.title('Correlation Heat Map between Features and Chrun')
plt.show()
```
